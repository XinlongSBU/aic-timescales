```python
def make_TRho_fig(ax, d):

    xs = d.log_center_Rho
    ys = d.log_center_T
    
    ax.plot(xs, ys, color='black')

    ax.set_xlabel(r'$\log(\rho_{\rm c}/\rm g \,cm^{-3})$')
    ax.set_xlim(9, 10)

    if ys.min() > 8:
        ymin = 8
    else:
        ymin = 7.5
    
    ax.set_ylabel(r'$\log(T_{\rm c}/\rm K)$')
    ax.set_ylim(ymin,9)
```

```python
def make_timescale_fig(ax, d):

    xs = d.log_center_Rho

    ys = np.log10(d.t_accretion)
    ax.plot(xs, ys, label=r'$t_{\rm accrete}$', ls='solid')

    ys = np.log10(d.center_t_compress)
    ax.plot(xs, ys, label=r'$t_{\rm compress}$', ls='dashed')

    ys = np.log10(d.center_t_capture)
    ax.plot(xs, ys, label=r'$t_{\rm capture}$', ls='dotted')

    ys = np.log10(np.abs(d.center_t_heat))
    ax.plot(xs, ys, label=r'$t_{\rm heat}$', ls='dashdot')


    ax.set_xlabel(r'$\log(\rho_{\rm c}/\rm g \,cm^{-3})$')
    ax.set_xlim(9, 10)

    ax.set_ylabel(r'$\log(t/\rm s)$')
    ax.set_ylim(0, 18)

    ax.legend()
```

```python
def make_combined_figure(d):
    fig, axs = plt.subplots(1,2)
    fig.set_size_inches(7,3.4)
    make_TRho_fig(axs[0], d)
    make_timescale_fig(axs[1], d)
    return fig
```

# Fiducial Model

```python
ld40 = mr.MesaLogDir('output/ONe6040/')
d40 = ld40.history
```

```python
def started_captures(center_ne20):
    return center_ne20 < 0.39
m_nums = ld40.select_models(started_captures, 'center_ne20')
```

```python
m_nums = [1164, 1173, 1182, 1192, 1201, 1210, 1219, 1228, 1237, 1265]
```

```python
fig, axs = plt.subplots(2,1, sharex=True)
for m in m_nums:
    d = ld40.profile_data(model_number=m)
    xs = d.logR+np.log10(7e10)
    axs[0].plot(xs, d.logT)
    axs[1].plot(xs, d.ne20)

axs[0].set_ylim(8.5, 9.05)
axs[0].set_ylabel(r'$\log(T/\rm K)$')
axs[1].set_ylim(0, 0.4)
axs[1].set_ylabel(r'$X_{\rm Ne}$')
axs[1].set_xlim(5, 8)
axs[1].set_xlabel(r'$\log(r/\rm cm)$')

fig.subplots_adjust(wspace=0, hspace=0.05)
fig.set_size_inches(3.4, 3.4)
fig.savefig('./notes/ONe6040-profiles.pdf')
```

```python
fig, ax = plt.subplots()

l1, = ax.plot(d40.log_center_Rho, d40.log_center_T, color='black', label='center', lw=1)
l2, = ax.plot(d40.max_eps_nuc_lgRho, d40.max_eps_nuc_lgT, color='black', ls='dotted', label=r'max $\epsilon_{\rm nuc}$')

for i, m in enumerate(m_nums):
    d = ld40.profile_data(model_number=m)
    ax.scatter(d.logRho[-1], d.logT[-1], color=f'C{i}')
    idx = d.eps_nuc.argmax()
    ax.scatter(d.logRho[idx], d.logT[idx], color=f'C{i}')

ax.set_xlabel(r'$\log(\rho/\rm g \,cm^{-3})$')
ax.set_xlim(9.7, 10)

ax.set_ylabel(r'$\log(T/\rm K)$')
ax.set_ylim(8.5,9.05)

ax.legend(loc=4)
fig.set_size_inches(3.4, 3.4)
fig.savefig('./notes/ONe6040-RhoT.pdf')
```
